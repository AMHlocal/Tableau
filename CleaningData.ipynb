# using iqr to find the upper outliers to remove

```python
quartiles = march['tripduration'].quantile([0.25, 0.5, 0.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of trip duration is: {lowerq} seconds")
print(f"The upper quartile of trip duration is: {upperq} seconds")
print(f"The interquartile range of trip duration is: {iqr} seconds")
print(f"The the median of trip duration is: {quartiles[0.5]} seconds")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} seconds could be outliers.")
print(f"Values above {upper_bound} seconds could be outliers.")
```
